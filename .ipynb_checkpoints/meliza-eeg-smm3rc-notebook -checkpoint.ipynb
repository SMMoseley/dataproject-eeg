```python
#Plotting multiple channels and their average for practice
#Selecting the channels to plot
datapractice=data_retrieval('resources\P1\P1_20190912_mismatch_r2.arf')
datalist=list(datapractice.keys())
print('List of datasets in this file:\n',datalist)
rec1=datapractice.get('rec_1')
lsrec1=list(rec1.keys())
print('List of datasets in this file:\n',lsrec1)
stuff= list(rec1)
all_channels = stuff[2:42]
channels = all_channels[0:5]

#Plotting the channels and their average
rec1_channels = []
for i in channels:
	rec1_channels.append(rec1[i][:])
	plt.plot(rec1[i][0:5000],color="blue",linewidth=0.2)
rec1_average = np.mean(rec1_channels,axis=0)
plt.plot(rec1_average[0:5000],color="darkorange",linewidth=0.5)
plt.xlabel("Time Sample")
plt.ylabel("mV")
plt.title("Sample EEG for Recording 1")
plt.show()
```

```python
#Loading channel data into MNE
sfreq = 2000  # Sampling frequency
included_channels=['channel0','channel1','channel2','channel3','channel4','channel5','channel6','channel7','channel8','channel9','channel10', 'channel11', 'channel12', 'channel13', 'channel14', 'channel15', 'channel16', 'channel17', 'channel18', 'channel19', 'channel20', 'channel21', 'channel22', 'channel23', 'channel24', 'channel25', 'channel26', 'channel27', 'channel28', 'channel29','channel30', 'channel31']
#Setting channel types
ch_types = ['eeg'] * len(included_channels)
#Stacking channel data into rows on top of each other
data = np.row_stack([rec_1[c][:] for c in included_channels])
#Loading into MNE
info = mne.create_info(ch_names=included_channels, sfreq=sfreq, ch_types=ch_types)
info
```

```python
# Plotting eeg data
raw = mne.io.RawArray(data, info)

# Scaling of the figure.
# For actual EEG/MEG data different scaling factors should be used.
scalings = 'auto'

raw.plot(n_channels=32, scalings=scalings, title='Data from arrays',
         show=True, block=True)
```
